# Matrix

```python
import copy

h, w = list(map(lambda x: int(x), input().split(" ")))

matrix = []

for i in range(h):
    matrix.append(list(input()))

def color_check(mtx, x, y):
    if x >= 0 and y >= 0 and y < h and x < w:
        if mtx[y][x] == type_:
            return True
    return False

def floodfill(mtx, x, y):
    queue = [[x,y]]
    mtx[y][x] = 'c'
    while queue != []:
        x1, y1 = queue.pop()
        if color_check(mtx, x1, y1 + 1):
            queue.append([x1, y1 + 1])
            mtx[y1 + 1][x1] = 'c'
        if color_check(mtx, x1, y1 - 1):
            queue.append([x1, y1 - 1])
            mtx[y1 - 1][x1] = 'c'
        if color_check(mtx, x1 + 1, y1):
            queue.append([x1 + 1, y1])
            mtx[y1][x1 + 1] = 'c'
        if color_check(mtx, x1 - 1, y1):
            queue.append([x1 - 1, y1])
            mtx[y1][x1 - 1] = 'c'

    return mtx

for i in range(int(input())):
    h1, w1, h2, w2 = list(map(lambda x: int(x) - 1, input().split(" ")))
    if matrix[h1][w1] != matrix[h2][w2]:
        print("neither")
    else:
        type_ = matrix[h1][w1]
        temp_mtx = floodfill(copy.deepcopy(matrix), w1, h1)
        if temp_mtx[h2][w2] == 'c':
            print("binary" if type_ == '0' else "decimal")   
        else: print('neither')
```

```python
h, w = list(map(lambda x: int(x), input().split(" ")))

matrix = []

for i in range(h):
    matrix.append(list(input()))

def color_check(x, y, colored):
    try:
        if matrix[y][x] == type_ and x >= 0 and y >= 0 and [x, y] not in colored:
            return True
    except:
        pass

    return False

def floodfill(x, y):
    queue = []
    colored = []
    queue.append([x, y])
    colored.append([x, y])
    while len(queue) > 0:
        x1, y1 = queue.pop()
        if color_check(x1, y1 + 1, colored):
            queue.append([x1, y1 + 1])
            colored.append([x1, y1 + 1])
        if color_check(x1, y1 - 1, colored):
            queue.append([x1, y1 - 1])
            colored.append([x1, y1 - 1])
        if color_check(x1 + 1, y1, colored):
            queue.append([x1 + 1, y1])
            colored.append([x1 + 1, y1])
        if color_check(x1 - 1, y1, colored):
            queue.append([x1 - 1, y1])
            colored.append([x1 - 1, y1])

    return colored

for i in range(int(input())):
    h1, w1, h2, w2 = list(map(lambda x: int(x), input().split(" ")))
    h1 -= 1
    w1 -= 1
    h2 -= 1
    w2 -= 1
    if matrix[h1][w1] != matrix[h2][w2]:
        print("neither")
    else:
        type_ = matrix[h1][w1]
        pos = floodfill(w1, h1)
        if [w2, h2] in pos:
            print("binary" if type_ == '0' else "decimal")   
        else: print('neither')
```
